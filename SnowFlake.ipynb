```python
import pandas as pd
import snowflake.connector

conn=snowflake.connector.connect(
    user='gowthamx25',
    password='Gowtham25052006',
    account='VFXKFNG-KQ94973',
    database='Eye',
    schema='PUBLIC',
    warehouse='COMPUTE_WH'
    )

table='select * from "EYE"'
df =pd.read_sql(table,conn)

conn.close()

print(df.head())
```
